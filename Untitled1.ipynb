```python
# Titanic Survival Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
df = pd.read_csv(url)

# Drop columns
df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Convert types
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')

# Create helper 
df['IsChild'] = df['Age'] < 18


df.head()

# Q1: Overall survival rate
print("Q1: Overall Survival Rate:")
print(f"{df['Survived'].mean():.2%}\n")

# Q2: Survival rate by gender
print("Q2: Survival Rate by Gender:")
print(df.groupby('Sex')['Survived'].mean().apply(lambda x: f"{x:.2%}"))
print()

# Q3: Survival rate by class
print("Q3: Survival Rate by Class:")
print(df.groupby('Pclass')['Survived'].mean().apply(lambda x: f"{x:.2%}"))
print()

# Q4: Average age of survivors vs non-survivors
print("Q4: Average Age of Survivors vs Non-Survivors:")
print(df.groupby('Survived')['Age'].mean().round(2))
print()

# Q5: Child vs Adult survival rate
print("Q5: Survival Rate - Children vs Adults:")
print(df.groupby('IsChild')['Survived'].mean().rename({True: "Children", False: "Adults"}).apply(lambda x: f"{x:.2%}"))
print()

# Q6: Survival rate by family aboard
print("Q6: Survival Rate by Family Aboard:")
print(df.groupby(df['SibSp'] > 0)['Survived'].mean().rename({True: "With Family", False: "Alone"}).apply(lambda x: f"{x:.2%}"))
print()

# Q7: Survival rate by embarkation port
print("Q7: Survival Rate by Embarkation Port:")
print(df.groupby('Embarked')['Survived'].mean().apply(lambda x: f"{x:.2%}"))
print()

# Q8: Average fare of survivors vs non-survivors
print("Q8: Average Fare - Survivors vs Non-Survivors:")
print(df.groupby('Survived')['Fare'].mean().round(2).rename({0: "Did Not Survive", 1: "Survived"}))

```

Exploring The Titanic Survival Odds

Data Prep:

I cleaned the dataset by dropping columns I did not need like 'Name' or 'Ticket' while also converting other types into catagories to help with structure and analysis.

I also created a new column for children to help compute survival odds.

I created 8 questions which I included below along with the data:

1. Only 38.8% of people survived showing how much of a tragedy this really was.
2. 74.2% of women survived which proves that this was a very important and influential factor when considering rescues. 
3. First class passangers had a survival rate of 62.9% while only 24.24% of third class passengers made it. A weathier person may have had a better chance at surviving this wreck.
4. Survivors were slightly younger being 28.34 while fatalities averaged at 30.63.
5. Children had a much higher chance of survival with 53.98% of them surviving.
6. Families were slightly more likely to survive having a survival rate of 46.64%.
7. Passengers who boarded in Cherbourg had the highest survival rate at 55.36%.
8. The average fare for survivors was much higher averaging at 48.4 while the other average was 22.12. This shows that being more wealthy may have benefitted some passangers.

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()

sns.histplot(data=df, x='Age', hue='Survived', multiple='stack')
plt.title('Survivors vs Non-Survivors by Age')
plt.show()
```
